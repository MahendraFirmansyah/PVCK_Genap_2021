```python
hasil = []

for x in range(0,img.shape[0]):
    row = []
    for y in range(0, img.shape[1]):
        blue = img[x,y,0] - 35
        green = img[x,y,1] - 35
        red = img[x,y,2] - 160
        blue = blue**2
        green = green**2
        red = red**2
        tambahRGB = blue + green + red
        akar = math.sqrt(tambahRGB)
        if akar > 55:
            row.append(0)
        else:
            row.append(1)
    hasil.append(row)

segar = True
for i in hasil:
    for j in i:
        if j > 0:
            segar = False
            break;
if segar:
    print("ikan Segar")
else:
    print("ikan Tidak Segar")
```
